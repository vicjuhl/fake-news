```python
import json
import pandas as pd
import pathlib as pl
import os
import numpy as np
import csv
```

```python
path = (pl.Path(os.path.abspath('')).parent.resolve() / "data_files/words/included_words_valset2.json")
with open(path) as input_file:
    # creating dataframe by reading json file directly
    data = json.load(input_file)
    n_articles = data["nArticles"] # unpack data
    words = data["words"]
# filtering for fake and reliable and replacing NaN with [0,0]
df = pd.DataFrame.from_dict(words, orient='index') #type: ignore
df = df.applymap(lambda x: [0,0] if x is np.nan else x)
df['freq'] = df.apply(
    lambda row: [row['reliable'][0] + row['fake'][0], row['reliable'][1] + row['fake'][1]], axis=1
) # adding freq entry column
df["freq"] = df["freq"].apply(lambda x: x[1])
df = df.sort_values(by='freq', ascending=False) # sort by total frequency
```

```python
df["fr_ratio"] = df["fake"].apply(lambda x: x[0]) / df["reliable"].apply(lambda x: x[0])
```
